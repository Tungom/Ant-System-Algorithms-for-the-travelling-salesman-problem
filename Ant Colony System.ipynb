```python
import ACOAs
import RTSP
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

# 1. Random Graph generator default parameters:
RandGraph(dim = 10, domain = (2,4))

```python
GRAPH = RTSP.RandGraph(20, (10,20))

# np.array(GRAPH)
g = nx.Graph() 

for i in range(len(GRAPH)):
    edges = [(i, k) for k,j in enumerate(GRAPH[i]) if j>0]
    for edge in edges:
        g.add_edge(edge[0], edge[1])

nx.draw(g, with_labels = True) 
```

```python
hm = plt.imshow(GRAPH, 'gray', origin='lower')

plt.colorbar(hm)
```

# 2. Ant Colony Sytem default parameters:
    
ACS(GRAPH, Population = 10, eps = 0.1, q0 = 0.9, alpha = 1, beta = 3, rho = 0.5, iterations = generations, dropout = False, show = False)

```python
Population = 10
eps = 0.1
q0 = 0.9
alpha = 1
beta = 3
rho = 0.1
generations = 500
dropout = False
show = False

for i in range(10):
    SOLUTION = ACOAs.ACS(GRAPH, Population = 10, eps = 0.1, q0 = 0.9, alpha = 1, beta = 3, rho = 0.5, iterations = generations,
                         dropout = False, show = False)
    print(SOLUTION["BSF"])
```

```python
hm = plt.imshow(SOLUTION["PRM"], 'gray', origin='lower')
plt.colorbar(hm)
```
